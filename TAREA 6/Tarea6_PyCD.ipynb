```python
import librosa
import librosa.display
import matplotlib.pyplot as plt
import numpy as np
import IPython
```

### Hola Helena

```python
IPython.display.Audio('/content/Hola Helena.mp4')
```

```python
h1, sr = librosa.load('/content/Hola Helena.mp4')
# h1: serie de tiempo de la frecuencia
# sr: frecuencia de muestreo
```

```python
plt.figure()
plt.plot(h1)
plt.xlabel('Muestras')
plt.ylabel('Amplitud')
plt.show()
```

```python
IPython.display.Audio(h1[13000:24000], rate= sr)
```

```python
h1 = h1[13000:24000]
```

```python
# Dividir entre frecuencia de muestreo da tiempo en segundos
x = [x / sr for x in range(len(h1))]
```

```python
plt.figure()
plt.plot(x, h1)
plt.xlabel('Tiempo (s)')
plt.ylabel('Amplitud')
plt.show()
```

```python
# Calcular MFCC
mfccs = librosa.feature.mfcc(y=h1, sr=sr, n_mfcc=13)

# Visualizar MFCC
plt.figure(figsize=(10, 4))
librosa.display.specshow(mfccs, x_axis='time')
plt.xlabel('Tiempo (s)')
plt.ylabel('Coeficientes MFCC')
plt.colorbar(format='%+2.0f dB')
plt.tight_layout()
plt.show()
```

```python
# Crear una figura con dos subgrficos
fig, ax = plt.subplots(nrows=2, sharex=True, figsize=(12, 8))

# Mostrar la forma de onda en el primer subgrfico
librosa.display.waveshow(h1, sr=sr, ax=ax[0])
ax[0].set(title='Forma de Onda')

# Mostrar los coeficientes MFCC en el segundo subgrfico
img = librosa.display.specshow(mfccs, x_axis='time', sr=sr, ax=ax[1])
ax[1].set(title='Coeficientes MFCC')

# Mostrar la figura
plt.show()
```

```python
# Calcular el espectrograma de la seal de audio
S = librosa.stft(h1)
S_db = librosa.amplitude_to_db(np.abs(S), ref=np.max)

# Visualizar el espectrograma
plt.figure(figsize=(10, 6))
librosa.display.specshow(S_db, sr=sr, x_axis='time', y_axis='log')
plt.colorbar(format='%+2.0f dB')
plt.title('Espectrograma')
plt.xlabel('Tiempo (s)')
plt.ylabel('Frecuencia (Hz)')
plt.show()
```

```python
# Obtener las frecuencias y los tiempos del espectrograma
frequencies = librosa.fft_frequencies(sr=sr)
len(frequencies)
```

```python
times = librosa.frames_to_time(np.arange(S_db.shape[1]), sr=sr)

# Sumar las amplitudes a lo largo del tiempo para cada frecuencia
sum_amplitudes = np.sum(S_db, axis=1)

# Seleccionar las 10 frecuencias ms representativas
top_n = 10
top_indices = np.argsort(sum_amplitudes)[-top_n:][::-1]
top_frequencies = frequencies[top_indices]
top_amplitudes = sum_amplitudes[top_indices]
top_amplitudes
```

```python
# Visualizar las frecuencias representativas como lneas
plt.figure(figsize=(10, 6))
plt.stem(top_frequencies, top_amplitudes, linefmt='b-', markerfmt='bo', basefmt='r-')
plt.xlabel('Frecuencia (Hz)')
plt.ylabel('Amplitud Total (dB)')
plt.title('Frecuencias Representativas del Espectrograma')
plt.show()
```
