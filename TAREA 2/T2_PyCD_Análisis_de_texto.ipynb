```python
def extraer_libro(FILE):
  fin = open(FILE, 'rb')
  lines = []
  for line in fin:
      line = line.strip().lower()
      line = line.decode("utf-8", "ignore")
      if len(line) == 0:
          continue
      lines.append(line)
  fin.close()
  return lines
```

```python
def tokensf(txt):
    texto_limpio = re.sub(r'[^\w\s]', '', txt.lower())
    #definir stop words
    stop_words = set(stopwords.words('spanish'))
    ## agregamos palabras a las stop_words despues del analizis inicial
    stop_words.add('tan')
    stop_words.add('si')
    stop_words.add('así')
    stop_words.add('usted')
    stop_words.add('dijo')
    stop_words.add('sólo')
    stop_words.add('ser')
    stop_words.add('mismo')
    stop_words.add('sido')
    tokens = word_tokenize(texto_limpio)
    texto_filtrado = [word for word in tokens if word not in stop_words]
    return texto_filtrado
```

```python
def get_tokens(map_libro):
  tokens = []
  for token in map_libro:
    tokens.extend(token)
  return tokens
```

```python
## Dividir por capitulo
def dividir_caps(lista, sep):
  series_text = pd.Series(lista)
  id_caps = series_text.str.contains(sep, regex=True)
  id_caps = series_text[id_caps].loc[0:]
  idx = id_caps.index
  caps= []
  rangos = []
  for i in range(len(idx)-1):
    caps.append(series_text.iloc[id_caps.index[i]:id_caps.index[i+1]])
    rangos.append((id_caps.index[i], id_caps.index[i+1]))
  return id_caps, caps, rangos
```
