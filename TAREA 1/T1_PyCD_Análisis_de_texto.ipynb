```python
def extraer_libro(FILE):
  fin = open(FILE, 'rb')
  lines = []
  for line in fin:
      line = line.strip().lower()
      line = line.decode("utf-8", "ignore")
      if len(line) == 0:
          continue
      lines.append(line)
  fin.close()
  return lines
```

```python
def tokensf(txt):
    texto_limpio = re.sub(r'[^\w\s]', '', txt.lower())
    #definir stop words
    stop_words = set(stopwords.words('spanish'))
    ## agregamos palabras a las stop_words despues del analizis inicial
    stop_words.add('tan')
    stop_words.add('si')
    stop_words.add('así')
    stop_words.add('usted')
    stop_words.add('dijo')
    stop_words.add('sólo')
    stop_words.add('ser')
    stop_words.add('mismo')
    stop_words.add('sido')
    tokens = word_tokenize(texto_limpio)
    texto_filtrado = [word for word in tokens if word not in stop_words]
    return texto_filtrado
```

```python
def get_tokens(map_libro):
  tokens = []
  for token in map_libro:
    tokens.extend(token)
  return tokens
```

```python
comunes = set(df_frecuencias1.head(20)['token']).intersection(df_frecuencias2.head(20)['token'], df_frecuencias3.head(20)['token'])

data = []
for palabra in comunes:
    f1 = df_frecuencias1.loc[df_frecuencias1['token'] == palabra, 'conteo'].iloc[0]
    data.append((palabra, f1, "Ana María Rodríguez"))
    f2 = df_frecuencias2.loc[df_frecuencias2['token'] == palabra, 'conteo'].iloc[0]
    data.append((palabra, f2, "Marta Salís"))
    f3 = df_frecuencias3.loc[df_frecuencias3['token'] == palabra, 'conteo'].iloc[0]
    data.append((palabra, f3, "Lourdes Íñiguez"))

df = pd.DataFrame(data, columns=["Palabra", "Frecuencia", "Traducción"])

plt.figure(figsize=(12, 6))
sns.set_palette("mako")
sns.barplot(data=df, x="Palabra", y="Frecuencia", hue="Traducción")
plt.xlabel('Palabra')
plt.ylabel('Frecuencia')
plt.xticks(rotation=45)
plt.legend(loc=0)
plt.tight_layout()
plt.show()
```

```python
comunes = set(df_frecuencias1.head(20)['token']).intersection(df_frecuencias2.head(20)['token'], df_frecuencias3.head(20)['token'])

data = []
for palabra in comunes:
    f1 = df_frecuencias1.loc[df_frecuencias1['token'] == palabra, 'conteo'].iloc[0]/df_frecuencias1['conteo'].sum()
    data.append((palabra, f1*100, "Ana María Rodríguez"))
    f2 = df_frecuencias2.loc[df_frecuencias2['token'] == palabra, 'conteo'].iloc[0]/df_frecuencias2['conteo'].sum()
    data.append((palabra, f2*100, "Marta Salís"))
    f3 = df_frecuencias3.loc[df_frecuencias3['token'] == palabra, 'conteo'].iloc[0]/df_frecuencias3['conteo'].sum()
    data.append((palabra, f3*100, "Lourdes Íñiguez"))

df = pd.DataFrame(data, columns=["Palabra", "Frecuencia", "Traducción"])

plt.figure(figsize=(12, 6))
sns.set_palette("mako")
sns.barplot(data=df, x="Palabra", y="Frecuencia", hue="Traducción")
plt.xlabel('Palabra')
plt.ylabel('Frecuencia (%)')
plt.xticks(rotation=45)
plt.legend(loc=0)
plt.tight_layout()
plt.show()
```
